```python
import pandas as pd

# Assuming the csv file is named 'netflix_titles.csv' based on the zip file name
csv_file_path = os.path.join("netflix_data", "netflix_titles.csv")

# Read the csv file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame to confirm it loaded correctly
print(df.head())
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
df['year_added'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title("Content Added to Netflix Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns

heat_df = df.pivot_table(index='year_added', columns='month_added', values='show_id', aggfunc='count')

plt.figure(figsize=(12,6))
sns.heatmap(heat_df, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title("Netflix Content Added Over Time (Year vs Month)")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()
```
